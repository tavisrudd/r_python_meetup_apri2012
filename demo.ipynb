Vancouver bike lane volumes for 2010/2011:

Loading the Data with Pandas----------------------------

```python
weather_initial = pandas.read_csv('weather2010n.csv', parse_dates=True, index_col=0)
```

Check which columns are available:

```python
weather_initial.columns
```

Keep columns of interest:

```python
weather_columns = ['Total Rain (mm)', 'Total Snow (cm)', 'Min Temp (C)', 'Max Temp (C)']weather = weather_initial[weather_columns]weather
```

Load cycling data from Excel file

```python
cycling_xls = pandas.ExcelFile('cycling.xls')cycling_initial = cycling_xls.parse('A', header=2, index_col=0)cycling_initial
```

The column names aren't quite right (they were split over two rows in the .xls), so we rename them.

```python
cycling = cycling_initial.rename(columns={'Unnamed: 2': 'Burrard Bridge', 'Unnamed: 12': 'Dunsmir Viaduct'})cycling
```

Combining the datasets & picking the interesting bits-----------------------------------------------------

Join the cycling and weather data in to one combined DataFrame. Pandas joins on the index of each dataset.

```python
combined = cycling.join(weather, how='inner')combined
```

Only the 'Burrard Bridge' and 'Dunsmir Viaduct' routes have a good amount of data for the year. We extract just their columns and combine with the weather columns.

```python
route_columns = ['Burrard Bridge', 'Dunsmir Viaduct']interesting = combined[route_columns + weather_columns]interesting
```

Dealing with outliers---------------------

Find the index of the outlying datapoint

```python
interesting.xs(interesting['Burrard Bridge'].idxmax())
```

```python
interesting['Burrard Bridge'].describe()
```

Remove the outlier by replacing it with None. Pandas deals with this as "missing data".

```python
interesting.set_value(interesting['Burrard Bridge'].idxmax(), 'Burrard Bridge', None)print interesting['Burrard Bridge'][pandas.DateRange('2010/06/29', periods=1)]interesting['Burrard Bridge'].describe()
```

Missing data points can be interpolated when required.

```python
late_june = pandas.DateRange('2010/06/25', '2010/07/05')interesting['Burrard Bridge'][late_june].plot(figsize=figsize)
```

```python
interesting['Burrard Bridge'][late_june].interpolate().plot(figsize=figsize)
```

Finding trends in the data--------------------------

A quick look at all the data plotted in a matrix to visually identify trends or clusters.

```python
import pandas.tools.plottingsm = pandas.tools.plotting.scatter_matrix(interesting, figsize=figsize)
```

That's a lot of graphs. Let's focus on just the burrard bridge, total rain, and maximum temperature.

```python
burrard = interesting[['Burrard Bridge', 'Total Rain (mm)', 'Max Temp (C)']]sm2 = pandas.tools.plotting.scatter_matrix(burrard, figsize=figsize)
```

Let's get some numbers by computing the covariance of the data.

```python
burrard.cov()
```

Now the correlation. The default correlation method is Pearson correlation which assumes the data is linearly correlated.

```python
burrard.corr()
```

The Spearman correlation just assumes a monotonic correlation between the variables.In this case it shows a stronger negative correlation between trips on the bridge and the amount of rainfall.

```python
burrard.corr(method='spearman')
```

Comparing weekend and weekday activity--------------------------------------

The DateRange object is a powerful way to index time-series data

```python
weekdays = pandas.DateRange(start='2010/01/01', end='2010/12/31', time_rule='WEEKDAY')weekday_activity = burrard.ix[weekdays]weekday_activity.describe()
```

You can do math with DateRange indexes

```python
weekend_activity = burrard.ix[burrard.index - weekday_activity.index]weekend_activity.describe()
```

Visualizing the effects of rain-------------------------------

```python
norain = burrard.ix[burrard['Total Rain (mm)'] == 0]norain['Burrard Bridge'].hist()
```

```python
anyrain = burrard.ix[burrard['Total Rain (mm)'] > 0]anyrain['Burrard Bridge'].hist()
```

```python
morerain = burrard.ix[burrard['Total Rain (mm)'] > 10]morerain['Burrard Bridge'].hist()
```
